```python
flipkart_url="https://www.flipkart.com/search?q=sarees&sid=clo%2C8on%2Czpd%2C9og&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=sarees%7CWomen%27s+Sarees&requestId=f241db9f-4db4-4a97-b803-20e7e99d7723&as-backfill=on"
wanted_list=['Printed Kanjivaram Cotton Silk Saree','₹249']
```

```python
scraper=AutoScraper() # i have saved the sub package in the scraper variable
```

```python
result=scraper.build(flipkart_url,wanted_list) #putting the url and wanted list in the scraper build
```
